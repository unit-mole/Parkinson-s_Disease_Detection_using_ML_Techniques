## Model Building:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Separate features and target variable
X = df.drop(['name', 'status'], axis=1)
y = df['status']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Standardize the features:

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Logistic Regression:

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Logistic Regression
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
y_pred_train_lr = lr_model.predict(X_train)
y_pred_test_lr = lr_model.predict(X_test)
```

```python
# Training Accuracy
train_accuracy = accuracy_score(y_train, y_pred_train_lr)
```

```python
# Testing Accuracy
test_accuracy = accuracy_score(y_test, y_pred_test_lr)
```

```python
print("Logistic Regression:")
print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)
```

```python
print("\nClassification Report - Testing Data:")
print(classification_report(y_test, y_pred_test_lr))
```

```python
print("\nClassification Report - Training Data:")
print(classification_report(y_train, y_pred_train_lr))
```

```python
print("\nConfusion Matrix - Testing Data:")
print(confusion_matrix(y_test, y_pred_test_lr))
```

```python
print("\nConfusion Matrix - Training Data:")
print(confusion_matrix(y_train, y_pred_train_lr))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Get the predicted probabilities for both training and testing data
y_train_probs = lr_model.predict_proba(X_train)[:, 1]
y_test_probs = lr_model.predict_proba(X_test)[:, 1]

# Calculate the ROC curve and AUC score for training data
fpr_train, tpr_train, thresholds_train = roc_curve(y_train, y_train_probs)
roc_auc_train = roc_auc_score(y_train, y_train_probs)

# Calculate the ROC curve and AUC score for testing data
fpr_test, tpr_test, thresholds_test = roc_curve(y_test, y_test_probs)
roc_auc_test = roc_auc_score(y_test, y_test_probs)

# Plot the ROC curve for training and testing data
plt.figure(figsize=(8, 6))
plt.plot(fpr_train, tpr_train, label=f'Training ROC Curve (AUC = {roc_auc_train:.2f})')
plt.plot(fpr_test, tpr_test, label=f'Testing ROC Curve (AUC = {roc_auc_test:.2f})')
plt.plot([0, 1], [0, 1], 'k--', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Logistic Regression')
plt.legend()
plt.show()
```
